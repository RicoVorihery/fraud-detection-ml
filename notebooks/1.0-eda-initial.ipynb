```python
import pandas as pd
```

```python
df_transactions = pd.read_csv('../data/raw/train_transaction.csv')
df_identity = pd.read_csv('../data/raw/train_identity.csv')
```

```python
df_transactions.head(10)
```

```python
df_identity.head()
```

```python
df = df_transactions.merge(df_identity,on='TransactionID', how='left')
```

```python
df.head(10)
```

```python
df.shape #nombre de lignes x colonnes
```

```python
df.dtypes.head(20)
```

```python
#pourcentage de valeurs manquantes pour chaque colonne
df.isna().mean().sort_values(ascending=False) 
```

```python
#proportion de valeurs 0 (transaction normale) et 1 (fraude)
df['isFraud'].value_counts(normalize=True) 
```

```python
#Repérer les 5 features numériques les plus corrélées avec isFraud
df.corr(numeric_only=True)['isFraud'].abs().sort_values(ascending=False).head(6)
```
